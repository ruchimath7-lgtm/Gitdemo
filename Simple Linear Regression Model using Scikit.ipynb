## Predictive Analytics Project: Forecasting Sunday Circulation from Daily Circulation using Machine Learning (Linear Regression)

```python
import pandas as pd 
dc=pd.read_csv("C:\\Users\\Ruchi Goyal\\Downloads\\NewspaperData.csv")
dc.head(10)
```

## Checking for the linear regression between daily and sunday

```python
import matplotlib.pyplot as plt
plt.scatter(dc['daily'],dc['sunday'],color='r')
plt.xlabel=("daily")
plt.ylabel=("sunday")
```

## This shows that there is a linear regression between daily and sunday

```python
dc.corr()
```

## It shows that there is a postive correlation between 'daily' and 'sunday' as the value of correlation lies near to 1. 

```python
X=dc[['daily']]
Y=dc['sunday']
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,Y)
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

## Linear regression line equation denoting "daily" as 'd' and "sunday" as 's'.

```python
print(f"Linear regression line equation : s={round(model.intercept_,3)}d+{round(model.coef_[0],3)}")
```

## Predicting Sunday paper circulation when daily circulation is 200

```python
print("value of sunday sales when daily is :",200*model.coef_+model.intercept_)
```

## Predicting Sunday paper circulation when daily circulation is 200,300,500 using a trained model

```python
import pandas as pd
n_daily=[200,300,400]
n_model=pd.DataFrame({'n_daily':n_daily})
print(n_model)
```

```python

n_sunday= model.predict(n_model)
```

```python

print(pd.DataFrame({'n_daily':n_daily,'n_sunday' : n_sunday}))
```
