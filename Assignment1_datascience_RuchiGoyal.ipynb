```python
ems={101:{'name': 'Satya', 'age': 27, 'department': 'HR', 'salary': 50000}}
def add_employee():
    emp_id=int(input("Enter the emp_id:"))
    if emp_id in ems:
        print("employee already exists")
    else:
        add_name=input("Enter the name:")
        add_age=int(input("Enter the age:"))
        add_dept=int(input("Enter the department:"))
        add_salary=int(input("Enter the salary:"))
        ems[emp_id]={'name': add_name, 'age': add_age, 'department': add_dept, 'salary': add_salary}
        print("Employee added successfully!")
def view_all():
    for emp_id, details in ems.items():
        print(f"ID: {emp_id}")
        print(f"  Name: {details['name']}")
        print(f"  Age: {details['age']}")
        print(f"  Department: {details['department']}")
        print(f"  Salary: {details['salary']}")
        print()
def search():
    d=int(input("Enter the employee id :"))
    if d in ems:
        print("Name:",ems[d]['name'])
        print("Deptt:",ems[d]['department'])
        print("Age:",ems[d]['age'])
        print("Salary:",ems[d]['salary'])
    else:
        print("Employee not found.")

def main_menu():
    a=int(input("Enter 1 to add new employee, enter 2 to view all employee, enter 3 to search for employee, enter 4 to exit:"))
    if a==1:
        add_employee()
    elif a==2:
        view_all()
    elif a==3:
        search()
    else:
        print("Thankyou")

main_menu()
```
