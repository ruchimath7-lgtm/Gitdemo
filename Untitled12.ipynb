### Now calculating VIF to check which column to drop

### from statsmodels.stats.outliers_influence import variation_inflation_factor
### from statsmodels.tools.tools import add_constant

```python
##Since we can combine these two columns thus adding a new columns which is 
##balanceOrg="newbalanceOrg"-"oldbalanceOrg" 
##balanceDest= "newbalanceDest" - "oldbalanceDest"
```

```python
data["balanceOrg"]=data["newbalanceOrig"]-data["oldbalanceOrg"] 
data["balanceDest"]=data["newbalanceDest"]-data["oldbalanceDest"] 
data.head()
```

```python
df=data.drop(["nameOrig","oldbalanceOrg","newbalanceOrig","nameDest","oldbalanceDest","newbalanceDest"],axis=1)
df.head()
```

```python
df.corr()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df[['amount', 'balanceDest']]

vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
print(vif_data)
```

## Since the VIF < 5, thus there is no multicollinearity now in the data.

```python
##Now checking for outliers 
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))
print("Number of outliers per column:")
print(outliers_iqr.sum())
```
