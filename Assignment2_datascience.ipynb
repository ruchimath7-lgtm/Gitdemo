  ## b. Compute the dot product of the two arrays. 


```python
arr1=np.random.randint(1,20,size=(3,3))
arr2=np.random.randint(1,20,size=(3,3))
print("First array is",arr1)
print("Second array is",arr2)
print("Element-wise addition:",arr1+arr2)
print("Element-wise subtraction:",arr1-arr2)
print("Element-wise Multiplication :",arr1*arr2)
print("Dot product of the two arrays",np.dot(arr1,arr2))
```

### 1. Create a simple line plot for the following data:  x = [1, 2, 3, 4, 5],y = [10, 15, 25, 30, 50]
###  a. Plot the data.  
###  b. Customize the plot by adding a title, axis labels, and a grid.  

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10, 15, 25, 30, 50]
plt.plot(x,y)   
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title("Line chart")   
plt.grid(True,linestyle='--')
plt.show()
```

## 3. Create a pie chart to represent the percentage distribution of a companys revenue from different regions:  
###  a. Create a pie chart with the region names as labels.  
### b. Highlight the region with the highest revenue.  


```python
regions=['North_America', 'Europe', 'Asia', 'Others']
revenue=[45, 25, 20, 10]
max_index = np.argmax(revenue)
explode = [0] * len(revenue)
explode[max_index] = 0.1   
plt.pie(revenue,labels=regions,explode=explode,autopct='%1.0f%%')  
plt.title("Region-wise revenue analysis") 
plt.show()
```

### 4. Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000).  


```python
plt.hist(np.random.randint(1,100,1000),color='g')
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.title("Frequency distribution of randomly generated integers")
plt.grid(True)
```
