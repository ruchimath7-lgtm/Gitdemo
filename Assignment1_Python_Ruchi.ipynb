```python
p=int(input("Enter the first number: ")) #Takes two numbers as input from the user.
q=int(input("Enter the second number: "))

print("Addition:",p+q) #Performs the basic mathematical operations on these two numbers and  display the results of each operation on the screen.
print("Subtraction:",p-q)
print("Multiplication:",p*q)
print("Division:",p/q)
```
